```python
f = open('2020.json',)
dados_2020 = json.load(f)
```

```python
f = open('2021.json',)
dados_2021 = json.load(f)
```

```python
columns = ['data_pregao',
        'cod_bdi',
        'cod_papel',
        'tp_merc',
        'nome_resum',
        'espec_papel',
        'preco_abertura',
        'preco_maximo',
        'preco_medio',
        'preco_minimo',
        'preco_ultimo',
        'preco_melhor_compra',
        'preco_melhor_venda',
        'num_negocios',
        'fator_cotacao',
        'prazo_dias_termo',
        'qtde_titulos',
        'vol_titulos',
        'preco_exerc',
        'data_vencimento',
        'moeda',
        'indicador_correcao',
        'preco_exerc_pontos',
        'cod_isi',
        'distribuicao_papel',]
```

```python
codigos = []

for x in dados_2020:
    codigos.append(x['cod_papel'])
    
codigos = list(set(codigos))
```

```python
dicionario_opcoes_compra = {}
```

```python
for codigo in tqdm.tqdm(codigos):
    
    if not 'AZUL' in codigo: continue
    
    dias_de_negociacao = []

    for x in dados_2020:
        if x['cod_papel'] == codigo:
            dias_de_negociacao.append(x)
            
        
    a_dictionary = {}
    for name in columns:
        a_dictionary[name] = []
        
    for x in dias_de_negociacao:
        for name in columns:
            if name == 'data_pregao':
                data = datetime.datetime.fromtimestamp(x['data_pregao']['$date'] / 1e3, pytz.timezone("Etc/GMT0"))
                a_dictionary[name].append(data)
            elif name == 'data_vencimento':
                data = datetime.datetime.fromtimestamp(x['data_vencimento']['$date'] / 1e3, pytz.timezone("Etc/GMT0"))
                a_dictionary[name].append(data)
            else:
                a_dictionary[name].append(x[name])
                
    df = pd.DataFrame()
    for name in columns:
        df[name] = a_dictionary[name]
        
    dicionario_opcoes_compra[codigo] = df
```
